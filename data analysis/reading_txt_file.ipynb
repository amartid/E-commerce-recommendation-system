```python
import pandas as pd

# Read in the .txt file and separate the values by ","
df = pd.read_csv("/home/mai22042/otto/src/out/test_output.txt",nrows=100000, header=None, sep=",")
```

```python
from tqdm import tqdm
# Create an empty dictionary to store the results
results = {}

# Get all unique values of df[0]
unique_values = df[0].unique()

# Iterate over the unique values of df[0]
for value in tqdm(unique_values):
    # Filter the rows where the first item is equal to the current value
    sub_df = df[df[0] == value]
    
    # Sort the dataframe by the last value on each line from biggest to smallest
    sub_df = sub_df.sort_values(by=5, ascending=False).head(10)
    
    # Extract the second item on each sorted line
    second_items = sub_df[1].tolist()
    
    # Add the second items to the results dictionary with the current value as the key
    results[value] = second_items
```

```python
print(results)
```

```python
import pandas as pd
from tqdm import tqdm
# Create an empty dictionary to store the results
results = {}

# Get all unique values of df[0]
unique_values = df[0].unique()

# Iterate over the unique values of df[0]
for value in tqdm(unique_values):
    # Filter the rows where the first item is equal to the current value
    sub_df = df.groupby(df[0] == value)
    # Sort the dataframe by the last value on each line from biggest to smallest
    sort_df1 = lambda x: x.sort_values(by=5, ascending=False).head(10)[1]
    
    # Extract the second item on each sorted line
    result = sub_df.apply(sort_df1).tolist()

    # Add the second items to the results dictionary with the current value as the key
    results[value] = result
```

```python
print(results)
```
